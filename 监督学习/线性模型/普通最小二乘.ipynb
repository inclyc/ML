# 线性回归模型

这些模型可以找到一个目标是特征的线性组合的回归模型。

数学上：

$$
\hat{y} (w, x) = w_0 + w_1 x_1 + w_2 x_2 + \cdots + w_p x_p 
$$

这里的回归任务通常不是广泛意义下的“分类”任务，分类任务可以使用Logistic回归完成。

# 普通最小二乘模型

最小二乘法：最小化一个参数矢量 $(w_1, w_2, w_3, \cdots,w_p)$使得回归目标(y)与特征(x)的误差的平方最小。

也就是

$$
\min ||\textbf{X}w - y||^2
$$


注意，我们实际上还是会计算常数项$w_0$，只是为了简单起见没有写在这里

```python

import numpy as np
from numpy.random import rand
from matplotlib import pyplot as plt
size = 20
x = np.linspace(1, 5, size)
y = x * 0.5 + 3.2 + (rand(size, ) - 0.5) * 1.2

plt.scatter(x, y)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
reg = model.fit(x.reshape(size, 1), y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
yhat = x * reg.coef_ + reg.intercept_
plt.plot(x, yhat)
plt.scatter(x, y)
```

注意，这个算法的底层是SVD奇异值分解，复杂度是$O(n_{sample}{n_{feature}}^2)$
